```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
fixed_values = ["Stress_Level", "Depression_Score", "Anxiety_Score", "Financial_Stress"]

for col in data.columns:
    plt.figure(figsize=(10, 5))
    if data.dtypes[col] == "object" or col in fixed_values :
        sns.countplot(x=data[col])
    else:
        sns.kdeplot(data[col])
        plt.grid()
```

```python
data.groupby("Gender")["Anxiety_Score"].mean()
```

```python
data.groupby("Gender")["Depression_Score"].mean()
```

```python
for col in data.columns:
    plt.figure(figsize=(15, 5))
    if col != "Depression_Score":
        if data.dtypes[col] == "object" or col in fixed_values :
            sns.barplot(data.groupby(col)["Depression_Score"].mean(), palette="rainbow")
        else:
            sns.scatterplot(data.groupby(col)["Depression_Score"].mean())
            plt.grid()
        
    
    for elm in data[col].unique():
        print(f"{col}: {elm} -->", data[data[col] == elm]["Depression_Score"].mean())
```
